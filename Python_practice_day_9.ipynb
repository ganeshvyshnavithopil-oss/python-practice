```python
#Day 9
#Dictionary Items, Keys, Values
student = {
    "name": "Vyshu",
    "age": 21,
    "course": "MBA"
}

print("Dictionary:", student)

print("Keys:")
print(student.keys())

print("Values:")
print(student.values())

print("Items:")
print(student.items())
```

```python
#Capitalize dictionary value
data = {
    "city": "hyderabad",
    "country": "india"
}

capitalized = {}

for key, value in data.items():
    capitalized[key] = value.capitalize()

print(capitalized)
```

```python
#update dictionary value
student = {
    "name": "Ram",
    "age": 20
}

student.update({"course": "Python"})

print(student)
```

```python
#Square Root Program
import math

number = float(input("Enter a number: "))

result = math.sqrt(number)

print("Square root:", result)
```

```python
#Power and Square
number = int(input("Enter number: "))

square = math.pow(number, 2)
cube = math.pow(number, 3)

print("Square:", square)
print("Cube:", cube)
```

```python
#Factorial Using Loop
number = int(input("Enter number: "))

factorial = 1

for i in range(1, number + 1):
    factorial *= i

print("Factorial:", factorial)
```

```python
#Factorial Using Function
import math

number = int(input("Enter number: "))

print("Factorial:", math.factorial(number))
```

```python
#Trigonometry Programs
#Sin, Cos, Tan
#import math

angle = float(input("Enter angle in degrees: "))

radian = math.radians(angle)

print("Sin:", math.sin(radian))
print("Cos:", math.cos(radian))
print("Tan:", math.tan(radian))
```

```python
#Product of List Elements
numbers = [2, 3, 4, 5]

product = 1

for num in numbers:
    product *= num

print("Product:", product)
```

```python
#Product Using math.prod()
import math

numbers = [1, 2, 3, 4]

result = math.prod(numbers)

print("Product:", result)
```
