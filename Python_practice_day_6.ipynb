```python
#Day 6
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
num = int(input("Enter number: "))

if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")
```

```python
# for loops
for i in range(1, 11):
    print(i)
```

```python
num = int(input("Enter number: "))

for i in range(1, 11):
    print(num ,"×",i,"=", num * i)
```

```python
#while loop
i = 1

while i <= 5:
    print(i)
    i += 1
```

```python
#square condition
num = int(input("Enter number: "))

root = int(num ** 0.5)

if root * root == num:
    print("Perfect Square")
else:
    print("Not a Perfect Square")
```

```python
#def
def add(a, b):
    return a + b

result = add(5, 3)
print(result)
```

```python
def largest(a, b):
    if a > b:
        return a
    else:
        return b

print(largest(10, 20))
```

```python
#upper and lower case
text = input("Enter text: ")

print(text.upper())
```

```python
text = input("Enter text: ")

print(text.lower())
```

```python
text = input("Enter text: ")

count = 0

for ch in text:
    if ch.isupper():
        count += 1

print("Uppercase letters:", count)
```
